# ***Line plots***
## ***Using iris Data Set***

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set styles
# sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})


#load dataset
flower = sns.load_dataset("iris")
flower


#Change Figure
plt.figure(figsize=(8,6))


#Draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=flower)
plt.title("Iris flower data set")


# Adding Limit
# plt.xlim(2)
# plt.ylim(1)

# Set Styles
# - DarkGrid
# - WhiteGrid
# - Dark
# - White
# - Ticks

# sns.set_style("dark")


#showing plot
plt.show()
```
